```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
date_str = "2024.09.24"
time_str = "21.07.31"

cwd = os.getcwd() + f'\\SessionLogs\\{date_str}\\Athena\\{time_str}\\'


#path = cwd + 'SessionLogs/2024.08.25/Athena/20.23.32'
#path = Path.cwd() / 'data' / 'SessionLogs' / '2024.08.25' / 'Athena'  / '20.23.32'
df_motor = pd.read_csv(cwd + 'motor.csv')
df_imu = pd.read_csv(cwd + 'imu.csv').sort_values("time")

df_m0 = df_motor[df_motor['mid'] == 0]
df_m1 = df_motor[df_motor['mid'] == 1]
df_m2 = df_motor[df_motor['mid'] == 2]
df_m3 = df_motor[df_motor['mid'] == 3]
```

```python

# Create the figure and the axes
plt.figure(figsize=(10, 6))

# Plotting the three different line graphs
plt.plot(df_m0['time'], df_m0['output'], label='m0 output')
plt.plot(df_m1['time'], df_m1['output'], label='m1 output')
plt.plot(df_m2['time'], df_m2['output'], label='m2 output')
#plt.plot(df_m3['time'], df_m3['output'], label='m3 output')

#plt.plot(df_m0['time'], df_m0['angle'], label='m0 angle')
#plt.plot(df_m1['time'], df_m1['angle'], label='m1 angle')
#plt.plot(df_m2['time'], df_m2['angle'], label='m2 angle')

# Adding title and labels
plt.title('Motor Output')
plt.xlabel('x')
plt.ylabel('y')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
tmp = df_imu['time'].diff()

# Create the figure and the axes
plt.figure(figsize=(10, 6))

# Plotting the three different line graphs
plt.scatter(df_imu['time'], df_imu['time'].diff())
plt.plot(df_imu['time'], [tmp[tmp!=0].mean()]*df_imu['time'].count(), color='red')

# Adding title and labels
plt.title('timediff')
plt.xlabel('x')
plt.ylabel('y')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
df_acc = df_imu[df_imu['type'] == 'acceleration']
# Create the figure and the axes
plt.figure(figsize=(10, 6))

# Plotting the three different line graphs
plt.plot(df_acc['time'], df_acc['x'], label='x')
plt.plot(df_acc['time'], df_acc['y'], label='y')
plt.plot(df_acc['time'], df_acc['z'], label='z')

# Adding title and labels
plt.title('acceleration')
plt.xlabel('x')
plt.ylabel('y')

# Adding a legend
plt.legend()

# Show the plot
plt.show()

df_acc
```
